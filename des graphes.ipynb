```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

# Graphe 1

```python
plt.style.use('_mpl-gallery')

# make data
x = np.arange(50)
y = x*3

# plot
plt.figure(figsize=(5, 5))

plt.plot(x, y, color='r')

plt.grid()

plt.title('Ligne droite')

plt.xlabel('axe des x')
plt.ylabel('axe des y')

plt.show()
```

# Graphe 2    

```python
data = pd.read_csv('company_sales_data.csv', sep=',')
data
```

```python
x = data.month_number
y = data.total_profit

plt.figure(figsize=(5, 5))

plt.plot(x, y, linestyle='solid', color='b')
 
plt.grid()    

plt.title('Ventes')

plt.xlabel('mois')
plt.ylabel('profit')

plt.show()
```
