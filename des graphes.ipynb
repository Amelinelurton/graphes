```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Partie II

# Graphe 1

```python
plt.style.use('_mpl-gallery')

# make data
x = np.arange(50)
y = x*3

# plot
plt.figure(figsize=(15, 5))

plt.plot(x, y, color='r')

plt.grid()

plt.title('Ligne droite',fontsize=15)

plt.xlabel('axe des x',fontsize=13)
plt.ylabel('axe des y',fontsize=13)

plt.show()
```

# Graphe 2    

```python
data = pd.read_csv('company_sales_data.csv', sep=',')
data
```

```python
x = data.month_number
y = data.total_profit

plt.figure(figsize=(15, 5))

plt.plot(x, y, linestyle='solid', color='c', marker = 's', ms = 10, mfc = 'm')
 
plt.grid(axis = 'x')    

plt.title('Ventes',fontsize=15)

plt.xlabel('mois',fontsize=13)
plt.ylabel('profit',fontsize=13)

plt.show()
```

# Graphe 3

```python
#data

x = data.month_number
a = data.facecream
b = data.facewash
c = data.toothpaste
d = data.bathingsoap
e = data.shampoo
f = data.moisturizer

#plot

plt.figure(figsize=(15, 5))

plt.plot(x, a, linestyle='solid', color='b', label='facecream', marker = 'o')
plt.plot(x, b, linestyle='solid', color='m', label='facewash', marker = 'o')
plt.plot(x, c, linestyle='solid', color='g', label='toothpaste', marker = 'o')
plt.plot(x, d, linestyle='solid', color='r', label='bathingsoap', marker = 'o')
plt.plot(x, e, linestyle='solid', color='c', label='shampoo', marker = 'o')
plt.plot(x, f, linestyle='--', color='y', label='moisturizer', marker = 'o')

#présentation

plt.grid(axis = 'x')

plt.title('Ventes',fontsize=15)
plt.xlabel('mois')
plt.ylabel('produit')
plt.legend(loc='upper left',fontsize=11)


plt.show()
```

# Graphe 4

```python
data1 = data.drop(['total_units','total_profit'],axis = 1)
data1
```

```python
data1.plot.bar(x='month_number',figsize=(15,8))

plt.ylabel('unité produit vendu',fontsize=13)
plt.title('comparatif unité de produit vendu par catégorie',fontsize=30)

plt.legend(fontsize=16)

plt.grid(axis = 'x')

plt.show()
```

# Graphe 5

```python
data2 = data.drop(['total_units','total_profit','month_number'],axis = 1)
data2
```

```python
#définir liste produits
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

#définir liste des sommes
Sum =data2.sum()

#creating plot
plt.figure(figsize=(7, 7))

plt.axis("equal")
plt.pie(Sum, labels = products, autopct='%1.1f%%')

plt.title('Répartion des produits',fontsize=16)

plt.show()
```
